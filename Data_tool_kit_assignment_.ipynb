Question 2. How does broadcasting work in NumPy ?

Answer. Broadcasting allows NumPy to perform element-wise operations on arrays of different shapes without making unnecessary copies of data. It follows these rules:

If arrays have different dimensions, the smaller array is expanded to match the larger array.
If one array has a shape of (1, n) and another (m, n), NumPy replicates the smaller array along the missing dimension.
If shapes are incompatible after attempting to broadcast, an error is raised.

Question 17. Explain the difference between apply() and map() in Pandas ?

Answer. The apply() and map() functions in Pandas are used for data transformation, but they differ in their scope and functionality. The apply() function works on both DataFrames and Series, allowing the application of complex functions to entire rows or columns. It is useful when performing row-wise or column-wise operations in a DataFrame, such as aggregations or applying custom logic. In contrast, map() is limited to Series only and is used for element-wise transformations, making it ideal for modifying individual values or performing dictionary-based replacements. While apply() can handle complex functions and work on multiple columns, map() is simpler and more efficient for direct value modifications. For example, apply() is best suited for operations like computing row-wise sums in a DataFrame, whereas map() is useful for applying a function to each element in a Series, such as mapping categorical values to numerical labels.









Question 18.  What are some advanced features of NumPy ?

Answer. NumPy provides several advanced features that make it a powerful library for numerical computing and data manipulation. Some of its key advanced capabilities include:

Broadcasting – Enables operations between arrays of different shapes without explicit looping or reshaping, improving performance.
Vectorized Operations – Performs element-wise computations efficiently without using explicit Python loops, making operations significantly faster.
Memory Mapping – Allows working with large datasets by mapping files directly to memory, enabling efficient data processing without loading everything into RAM.
Structured Arrays – Supports heterogeneous data types within a single array, making it useful for handling tabular data efficiently.
Universal Functions (ufuncs) – Provides optimized mathematical functions that operate element-wise, such as np.sin(), np.exp(), and np.log().
Multi-dimensional Indexing – Allows advanced indexing techniques like boolean indexing, fancy indexing, and slicing for efficient data selection and modification.
Linear Algebra Operations – Offers a wide range of linear algebra functions, such as matrix multiplication (np.dot()), determinant calculation (np.linalg.det()), and eigenvalue decomposition (np.linalg.eig()).
Fourier Transform and Signal Processing – Supports Fast Fourier Transforms (FFT) for signal and image processing using np.fft.
Random Sampling & Probability Distributions – Provides functions for generating random numbers and working with probability distributions (np.random).
Parallel Computing with NumPy – Utilizes multi-threading and vectorization for performance optimization, especially in large-scale numerical computations.

Question 21. Why is NumPy’s array slicing faster than Python’s list slicing ?

Answer. NumPy’s array slicing is significantly faster than Python’s list slicing due to efficient memory management, contiguous storage, and optimized operations at the low-level. Unlike Python lists, which store references to objects, NumPy arrays use a contiguous block of memory with a fixed data type, allowing for fast retrieval and manipulation. When slicing a NumPy array, it creates a view (not a copy), meaning it doesn’t allocate new memory but instead references the original data. This makes slicing operations lightweight and efficient. In contrast, Python lists store elements as individual objects in different memory locations, and slicing a list creates a new copy, requiring extra memory and computational overhead. Additionally, NumPy is built on C and Fortran, utilizing vectorized operations and low-level optimizations, further boosting performance. These factors make NumPy’s array slicing much faster and more memory-efficient compared to Python lists, especially when working with large datasets.

```python
# Question 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap ?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a sample DataFrame
data = {
    'Age': [25, 30, 35, 40, 45],
    'Salary': [50000, 60000, 70000, 80000, 90000],
    'Experience': [2, 5, 7, 10, 15]
}

df = pd.DataFrame(data)

# Calculating the correlation matrix
corr_matrix = df.corr()

# Visualizing the correlation matrix with a heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Adding title
plt.title("Correlation Matrix Heatmap")

# Displaying the heatmap
plt.show()
```

```python
# Question 6. Create a DataFrame and add a new column based on an existing column ?

import pandas as pd

# Creating a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Salary': [50000, 60000, 70000, 80000]
}

df = pd.DataFrame(data)

# Adding a new column 'Bonus' (10% of Salary)
df['Bonus'] = df['Salary'] * 0.10

# Adding another column using a function
df['Total Compensation'] = df.apply(lambda row: row['Salary'] + row['Bonus'], axis=1)

# Displaying the DataFrame
print(df)
```

```python
# Question 7. Write a program to perform element-wise multiplication of two NumPy arrays ?

import numpy as np

# Creating two NumPy arrays
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([10, 20, 30, 40, 50])

# Element-wise multiplication using '*'
result1 = arr1 * arr2

# Element-wise multiplication using np.multiply()
result2 = np.multiply(arr1, arr2)

# Displaying the results
print("Array 1:", arr1)
print("Array 2:", arr2)
print("Result using '*':", result1)
print("Result using np.multiply():", result2)
```

```python
# Question 11. Perform matrix multiplication using NumPy ?

import numpy as np

# Creating two matrices (2x3 and 3x2)
A = np.array([[1, 2, 3],
              [4, 5, 6]])

B = np.array([[7, 8],
              [9, 10],
              [11, 12]])

# Matrix multiplication using '@' operator
result1 = A @ B

# Matrix multiplication using np.dot()
result2 = np.dot(A, B)

# Displaying the results
print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Result using '@':\n", result1)
print("Result using np.dot():\n", result2)
```

```python
# Question 13. Create a 3D scatter plot using Plotly ?

import plotly.graph_objects as go
import numpy as np

# Generating sample 3D data
np.random.seed(42)
x = np.random.randint(1, 100, 20)
y = np.random.randint(1, 100, 20)
z = np.random.randint(1, 100, 20)

# Creating the 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(size=6, color=z, colorscale='Viridis', opacity=0.8)
)])

# Adding labels and title
fig.update_layout(
    title="3D Scatter Plot using Plotly",
    scene=dict(
        xaxis_title="X Axis",
        yaxis_title="Y Axis",
        zaxis_title="Z Axis"
    )
)

# Display the plot
fig.show()
```
